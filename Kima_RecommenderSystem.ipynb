```python
import pandas
user_data=pandas.read_csv("movietitles.csv")
user_data
```

```python
ratings_data=pandas.read_csv("ratings.csv")
ratings_data
```

```python
# merge the data
data=pandas.merge(ratings_data,user_data,on ='item_id')
data
```

```python
# calculate the mean rating for each movie
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
# calculate the number of users who rated a paticular movie
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
# show the mean and count against movies
ratings=pandas.DataFrame(data.groupby('title')['rating'].mean())
ratings['num_of_ratings']=data.groupby('title')['rating'].count()
ratings.head(20)# head(20)----first twenty
```

```python
#pivot table to represent the users,movies and their ratings
pivot=data.pivot_table(index='user_Id',columns='title',values='rating')
pivot #pivot is a variable
```

```python
# calculate the correlation of the movies
selected_movie=pivot['Ace Ventura: Pet Detective (1994)']
similar=pivot.corrwith(selected_movie)
#similar is a var
similar_df=pandas.DataFrame(similar,columns=['correlations'])
similar_df.sort_values('correlations',ascending=False).head(20)
```

```python
#add num of ratings to the dataframe
similar_df=similar_df.join(ratings['num_of_ratings'])
similar_df[similar_df['num_of_ratings'] >100].sort_values('correlations',ascending=False).head(30)
```
