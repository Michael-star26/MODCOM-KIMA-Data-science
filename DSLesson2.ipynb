```python
import pandas
data=pandas.read_csv("data(1).csv")
data
# classification -uses categorical data
```

```python
# display the first 20
data.head(20)
```

```python
# divide the data into into x and y values
# x represents the features and y represents the outcome
array=data.values
X=array[:,0:8] #features/ingredients (flour...to...salt)
Y=array[:,8]# outcome (type)
```

```python
# split our dataset into train and test data
# use 70% of the data as train data and the other 30% as test data
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
# select the algorith to use
from sklearn.ensemble import RandomForestClassifier
# pip3 install -U scikit_learn
from sklearn.svm import SVC
# select the algorithm to use
from sklearn.ensemble import RandomForestClassifier
# pip install -U scikit_learn
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LinearRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# SVC HAS THE ACCURACY OF DATA USING SVC IS 93% WHILE RANDOMFORESTCLASSIFIER HAS ACCURACY OF 100%...SVC IS MUCH MORE ACCURATE

# cross validation- a method used to calculate the best algorithm to use
# use the training (70%=X_train,Y_train)
# create cluster(10)
# train and test the algorithms in each fold(cluster)
# get avarage accuracy score
# choose the best algorithm to use in our model
# fit the model in a list
models=[]
models.append(("Random Forest",RandomForestClassifier()))
models.append(("svc",SVC()))
models.append(("Guassian",GaussianNB()))
models.append(("Decision Tree Classifier",DecisionTreeClassifier()))
models.append(("K neighbours",KNeighborsClassifier()))
models.append(("Gradient Boosting",GradientBoostingClassifier()))
from sklearn.model_selection import cross_val_score,KFold

# use for loop to train and run all the algorithms
for name,model in models:
 #  define the number of folds/cluster
 kfold=KFold(n_splits=10,random_state=42,shuffle=True)
 #  get the result of each fold
 cv_results=cross_val_score(model,X_train,Y_train,cv=kfold,scoring='accuracy')
 # get the mean
 print("Model Name ",name,"Results: ",cv_results.mean())

model=SVC()

```

From the above results,two models namely DTC and Gradient boosting scored 100% (1.0) we can use one of the two as our models

```python
# fit the train data into the model
model.fit(X_train,Y_train)
```

```python
# use the model to predict X_test
prediction=model.predict(X_test)
print("The model predicted ",prediction)
print("Expected outcome ",Y_test)
```

```python
# check the accuracy score of the model
from sklearn.metrics import accuracy_score
print("The accuracy of the model is ",accuracy_score(Y_test,prediction))
```

```python
# predict using our own values
# flour milk sugar butter
recipe=[[50,25,5,8,9,3,1,0]]
outcome=model.predict(recipe)
print("I ampreparing ",outcome)
```

```python
# confussion matrix-checks where the model went wrong
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,prediction))
```

```python
# classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction))
```
