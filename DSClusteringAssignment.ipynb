```python
import pandas
data=pandas.read_csv("EastWestAirlines.csv")
data
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
# create a dataset of the columns to use in clustering
array=data.values
X=array[:,1:4]
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=8,random_state=42)
model.fit(X)
means=model.cluster_centers_ 
clusters=pandas.DataFrame(means,columns=['Bonus_miles','Flight_miles_12months','Flight_trans_12_months'])
clusters
```

```python
data['df']=model.labels_
cluster6=data[data['df']==5]
cluster6
```
