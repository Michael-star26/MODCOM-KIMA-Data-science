```python
# use visualization libraries to draw graphs
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.available
```

```python
plt.style.use("seaborn")
# draw countplot using seaborn
sns.countplot(x="Gender",data=data,palette="Reds")
plt.title("Distribution of Gender Status")
plt.xlabel("Gender")
plt.ylabel("Number Of members")
```

```python
sns.barplot(x="Gender",data=data,palette="magma",hue="Education",y="LoanAmount")
plt.title("Distribution Of gender Status")
plt.xlabel("Gender")
plt.ylabel("Loan Amount")
```

```python
# scatter plot
# hue---categorical data
sns.scatterplot(x="LoanAmount",y="ApplicantIncome",data=data,palette="magma",hue="Gender")
```

```python
# pie chart to represent the distribution of property area
data.groupby("Property_Area").size().plot(kind="pie",autopct="%1.1f%%")
plt.show()
```

Shows That majority of the people had property in semiUrban area at 37.9%
