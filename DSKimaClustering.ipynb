```python
import pandas
data=pandas.read_csv("Mall_Customers.csv")
data
```

```python
data.describe()
```

```python
# create a dataset of the columns to use in clustering
array=data.values
X=array[:,2:5] #age,Annual Income,Spending score
#subset=data[['Age','Annual Income','Spending Score']] #display specific data
#subset.head(10) #display the first 10
X
```

```python
# remove the inconsistent values
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python

array=data.values
X=array[:,2:5] 
```

```python
# K-means algorithm
from sklearn.cluster import KMeans
# define the number of clusters
# more clusters produce more accurate model 
model=KMeans(n_clusters=10,random_state=42)
# fit data into the model
model.fit(X)
```

```python
# check/display the 10 clusters created by KMeans
means=model.cluster_centers_  #find avarage mean
clusters=pandas.DataFrame(means,columns=['Age','Annual Income','Spending Score'])
clusters
```

members from cluster 2 are in their 40s,earning a lot of money but spending very little.The supermarket should motivate them and give them a better experience to improve their shoppinh

```python
# display members from the same cluster
data['df']=model.labels_
cluster6=data[data['df']==6]
cluster6
```

```python
# create a csv file with cluster 6 members
data[data['df']==6].to_csv("Cluster6.csv")
```
